````python
# Drop the 'id' and 'last_review' columns from the DataFrame 'df'
df.drop(['id', 'last_review'], axis=1, inplace=True)

# Display the DataFrame 'df' after dropping the columns
print("DataFrame after dropping 'id' and 'last_review' columns:")
df.head()
````

````python
# Fill the missing values in the 'name' column with 'Unknown'
df['name'].fillna('Unknown', inplace=True)

# Fill the missing values in the 'host_name' column with 'no_name'
df['host_name'].fillna('no_name', inplace=True)

# Fill the missing values in the 'reviews_per_month' column with 0
df.fillna({'reviews_per_month': 0}, inplace=True)
````

````python
# Check the number of remaining missing values from the columns we just filled
missing_names_count = df['name'].isnull().sum()
missing_host_names_count = df['host_name'].isnull().sum()
remaining_missing_values = df['reviews_per_month'].isnull().sum()

# Print the result
print("Number of missing values in the 'name' column:", missing_names_count)
print("Number of missing values in the 'host_name' column:", missing_host_names_count)
print("Number of remaining missing values in the 'reviews_per_month' column:", remaining_missing_values)
````

````python
# Review the DataFrame after filling the missing values
print("DataFrame after filling missing values:")
df
````

````python
# Check the number of properties with 0 availability throughout the year
round_the_year_busy_count = len(df[df['availability_365'] == 0])

# Print the result
print("Number of properties with 0 availability throughout the year:", round_the_year_busy_count)
````

````python
# Check the number of properties with a listing price of $0
properties_with_zero_price_count = len(df[df['price'] == 0])

# Print the result
print("Number of properties with a listing price of $0:", properties_with_zero_price_count)
````

````python
# Check the number of properties with a listing price between $1 and $10
properties_between_1_and_10_count = len(df[(df['price'] >= 1) & (df['price'] <= 10)])

# Print the result
print("Number of properties with a listing price between $1 and $10:", properties_between_1_and_10_count)
````

````python
# Calculate the number of properties with a listing price greater than or equal to $500
high_priced_properties_count = len(df[df['price'] >= 500])

# Print the result
print("Number of properties with a listing price of $500 or higher:", high_priced_properties_count)
````

````python
# Identify the rows with a listing price of $0 and get their index
index_names = df[df["price"] == 0].index
print("Indexes with listing price of $0:")
print(index_names)

# Drop the rows with index names corresponding to price $0 and store them in a variable
dropped_rows = df.drop(index_names)

# Print the indexes that have been dropped
print("\nIndexes that have been dropped:")
print(index_names)
````

````python
# Group the data by 'host_name', 'host_id', and 'neighbourhood_group', and find the maximum 'calculated_host_listings_count' for each host within their neighbourhood group.
hosts_listings = df.groupby(['host_name', 'host_id', 'neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()

# Sort the resulting DataFrame in descending order based on the 'calculated_host_listings_count', and select the top 10 hosts.
top_hosts = hosts_listings.sort_values(by='calculated_host_listings_count', ascending=False).head(10)

# Display the DataFrame showing the top hosts with the highest 'calculated_host_listings_count' within their neighbourhood group.
print("Top 10 Hosts With Highest Listings Count Within Neighbourhood Groups:")
top_hosts
````

````python
# Filter the DataFrame to select rows where the 'neighbourhood_group' is 'Manhattan' and the 'host_name' is 'Kara'
kara_listings_in_manhattan = df.loc[(df['neighbourhood_group'] == 'Manhattan') & (df['host_name'] == 'Kara')].head(5)

# Display the listings associated with Kara in Manhattan
print("Listings associated with Kara in Manhattan:")
kara_listings_in_manhattan
````

````python
# Chart - 12 visualization code
# Split the dataset by room type
entire_home_apt = df[df['room_type'] == 'Entire home/apt']
private_room = df[df['room_type'] == 'Private room']
shared_room = df[df['room_type'] == 'Shared room']

# Create boxplots for price
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].boxplot(entire_home_apt['price'])
axs[0].set_title('Entire home/apt')
axs[1].boxplot(private_room['price'])
axs[1].set_title('Private room')
axs[2].boxplot(shared_room['price'])
axs[2].set_title('Shared room')

# Add a common y-axis label
plt.ylabel('Price')

# Add a common title for the entire chart
fig.suptitle('Distribution Of Price By Room Type')

# Display the chart
plt.show()
````

This shows that price variable has many outliers in each ype of room, and this need to be removed.

````python
# Function to remove outliers from the 'price' variable
def remove_outliers(data):
    Q1 = np.percentile(data['price'], 25)
    Q3 = np.percentile(data['price'], 75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR
    data1 = data[(data['price'] >= lower_bound) & (data['price'] <= upper_bound)]
    return data1

# Remove outliers for each room type and create new subsets
entire_home_apt1 = remove_outliers(entire_home_apt)
private_room1 = remove_outliers(private_room)
shared_room1 = remove_outliers(shared_room)

# Combine the filtered subsets into a new DataFrame called 'combined_df'
combined_df = pd.concat([entire_home_apt1, private_room1, shared_room1], axis=0)
````

````python
# Create subplots with 1 row and 2 columns for two visualizations
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))
ax = axes.flatten()

# Show the final boxplot without outliers
sns.boxplot(x='room_type', y='price', data=combined_df, ax=ax[0])
ax[0].set_title('Price Distribution By Room Type (Outliers Removed)')
ax[0].set_xlabel('Room Type')
ax[0].set_ylabel('Price')

# Distplot gives us the univariate price distribution of observations
sns.distplot(combined_df['price'], ax=ax[1])
ax[1].set_title('Price Distribution (Outliers Removed)')
ax[1].set_xlabel('Price')
ax[1].set_ylabel('Density')

# Display the subplots
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

By using both boxplots and histograms with density plots, we can gain a comprehensive understanding of the price distribution for each room type and make comparisons between the room types. The visualizations help us identify potential outliers, understand the spread and central tendency of prices, and gain insights into the overall distribution of prices without the influence of extreme values.
